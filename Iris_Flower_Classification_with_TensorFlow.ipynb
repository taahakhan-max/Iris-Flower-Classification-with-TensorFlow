```python
train_path = tf.keras.utils.get_file(
    "iris training.csv","https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv"
)
test_path = tf.keras.utils.get_file(
    "iris test.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

train = pd.read_csv(train_path, names=CSV_COLUMN_NAMES, header=0)
test = pd.read_csv(test_path, names=CSV_COLUMN_NAMES, header=0)

train_y = train.pop('Species')
test_y = test.pop('Species')


def input_fn(features, labels, training=True, batch_size=256):
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    if training:
        dataset = dataset.shuffle(1000).repeat()

    return dataset.batch(batch_size)

# Normalize features (optional but helpful)
train = train / train.max()
test = test / test.max()

NUM_CLASSES = 3

#Build a DNN with 2 hidden layers with 30 and 10 hidden nodes each.
model  = tf.keras.Sequential([

      tf.keras.layers.Dense(30, activation='relu', input_shape=(train.shape[1],)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')
])


x = lambda:print('hi')
x()

# Compile the model BEFORE training
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # use 'categorical_crossentropy' if labels are one-hot
    metrics=['accuracy'])


# Train the Model.
model.fit(train, train_y, epochs=50, batch_size=32, validation_data=(test, test_y))


```

```python
# Evaluate the model on test data
eval_result = model.evaluate(test, test_y)

# Print test accuracy
print('\nTest set accuracy: {:.3f}\n'.format(eval_result[1]))
```
